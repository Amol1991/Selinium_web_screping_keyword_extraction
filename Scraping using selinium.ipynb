```python
import scrapy
import re
import time
from scrapy import Request
import logging
import json
from bs4 import BeautifulSoup
from lxml import etree
from scrapy.selector import Selector
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from lxml.etree import tostring
import pandas as pd
#https://www.geeksforgeeks.org/page_source-driver-method-selenium-python/
#https://www.geeksforgeeks.org/locator-strategies-selenium-python/
#https://www.geeksforgeeks.org/find_element_by_xpath-driver-method-selenium-python/?ref=rp

DRIVER_PATH = r"E:\downloads\chromedriver.exe"
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.get('https://www.amazon.com/Dove-Body-Wash-Pump-Moisture/product-reviews/B00MEDOY2G/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews')

reviewlist = []

#soup is one page and we are extracting info from that page.
def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:#we want body only.
            review = {'body': item.find('span', {'data-hook': 'review-body'}).text.strip()}#text.strip wil remove white spaces
            reviewlist.append(review)#here we will append the text in list
    except:
        pass
#I should mention that the webdriver will wait for a page to load by default. 
#It does not wait for loading inside frames or for ajax requests.
#It means when you use .get('url'), your browser will wait until the page is completely
#loaded and then go to the next command in the code. But when you are posting an ajax request, 
#webdriver does not wait and it's your responsibility to wait an appropriate amount of time for the 
#page or a part of page to load; so there is a module named expected_conditions.

for x in range(1,500):#initialize the loop
    driver.get('https://www.amazon.com/Dove-Body-Wash-Pump-Moisture/product-reviews/B00MEDOY2G/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={x}')
    element = WebDriverWait(driver, 5).until(
    EC.presence_of_element_located((By.XPATH, '//*[@class="a-section a-spacing-none review-views celwidget"]'))
    )
    time.sleep(1)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    get_reviews(soup)

with open('reviews.json','w') as fl:
    json.dump(reviewlist,fl)
```
